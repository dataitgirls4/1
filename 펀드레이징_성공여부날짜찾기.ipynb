```python
#general파일 불러온다
df1=pd.read_csv('data/general_dollar.csv',engine='python')
df1.head()
```

```python
#created date를 datetime으로 
df1['created_date'] =pd.to_datetime(df1['created_date'])
df1['created_date']
```

```python
#scrape date에서 date만 가져오기
df1['scrape_date'] =pd.to_datetime(df1['scrape_date'])
df1['scrape_date']=df1.scrape_date.dt.date
df1['scrape_date']=pd.to_datetime(df1.scrape_date)
df1['scrape_date'].head()
```

```python
#나중에 쓰기 위해 char_age 만들기
df1['char_age'] = df1['scrape_date']-df1['created_date']
df1['char_age']
```

```python
# timedelta인 char_age를 숫자형(float)으로 변환 (그래프 x축에 두기 위해)
df1['char_age'] = df1['char_age'].apply(lambda x:x.days)
df1.char_age
```

```python
#donation파일 불러온다
df2=pd.read_csv('data/donations_dollar.csv',engine='python')
df2.head()
```

```python
#donation table의 created_at을 datetime형으로 바꾸고 date만 뽑는다
df2['donated_at']=pd.to_datetime(df2.created_at)
df2['donated_at']=df2.donated_at.dt.date
df2['donated_at']=pd.to_datetime(df2.donated_at)
df2['donated_at'].head()
```

```python
#donation 너무 크니까 원하는 컬럼만 넣어서 뽑기
df2=df2[['chid','amount_dollar','donation_id','donated_at']]
df2.tail()
```

```python
#join 준비하기 df1의 chid도 df2처럼 float type으로 바꾸기
df1['chid'] = df1['chid'].astype(float)
df1.dtypes
```

```python
#general도 원하는 컬럼만 넣어서 뽑기
df1=df1[['chid','char_age','created_date','funds_target_dollar']]
df1.tail()
```

```python
# general과 donation join 하기
merge_1_2= pd.merge(df1, df2,
                  on='chid', how='inner')
merge_1_2
```

```python
#기부일자 - 모금프로그램생성일 = date_diff 컬럼 만들기
merge_1_2['date_diff'] = merge_1_2['donated_at']-merge_1_2['created_date']
merge_1_2
```

```python
# timedelta인 date_diff를 숫자형(float)으로 변환 (그래프 x축에 두기 위해)
merge_1_2['date_diff'] = merge_1_2['date_diff'].apply(lambda x:x.days)
merge_1_2.date_diff.dtypes
```

```python
#chid, date_diff로 groupby 하기
grouped = merge_1_2.groupby(['chid','date_diff'])['donation_id','amount_dollar','funds_target_dollar','char_age']
grouped
```

```python
#groupby 결과를 agg 함수로 보기
grouped=grouped.agg({'donation_id':'count','amount_dollar':'sum','funds_target_dollar':'mean','char_age':'mean'}).reset_index()
grouped
```

```python
#돈의 누적합 구하기
grouped['cumsum']=grouped.groupby(['chid'])['amount_dollar'].apply(lambda x:x.cumsum())
grouped
```

```python
#펀딩 성공여부 판단하는 success 컬럼 추가
grouped=grouped.assign(success = grouped['cumsum'] >= grouped['funds_target_dollar'])
grouped
```

```python
#success가 True일때, date_diff의 최소값과 프로그램 연식(?) 구하기
success_ch=grouped.query('success == True').groupby('chid').agg({'date_diff':'min','char_age':'mean'}).reset_index()
success_ch
```

```python
date_diff=success_ch['date_diff']
```

```python
date_diff.describe()
```

```python
#95% 신뢰구간 구하기
import scipy.stats as st

def mean_confidence_interval(date_diff, confidence=0.95):
    a = 1.0 * np.array(date_diff)
    n = len(a)
    m, se = np.mean(a), st.sem(a)
    h = se * st.t.ppf((1 + confidence) / 2., n-1)
    return m, m-h, m+h

mean_confidence_interval(date_diff)
```

```python
#17일 기준에서 성공프로그램 수, 배제되는 프로그램 수
# date-diff 컬럼을 선택
# 컬럼의 값과 조건을 비교
# 그 결과를 새로운 변수에 할당
successwithin_17 = success_ch['date_diff'] <= 17 


# char_age 컬럼을 선택합니다.
# 컬럼의 값과 조건을 비교합니다.
# 그 결과를 새로운 변수에 할당합니다.
olderthan_17 = success_ch['char_age'] >= 17

# 두가지 조건를 동시에 충족하는 데이터를 필터링하여 새로운 변수에 저장합니다. (AND)
date_diff17 = success_ch[successwithin_17 & olderthan_17]

# 결과를 출력합니다.
date_diff17
```

```python
#19일 기준에서 성공프로그램 수, 배제되는 프로그램 수
# date-diff 컬럼을 선택
# 컬럼의 값과 조건을 비교
# 그 결과를 새로운 변수에 할당
successwithin_19 = success_ch['date_diff'] <= 19 


# char_age 컬럼을 선택합니다.
# 컬럼의 값과 조건을 비교합니다.
# 그 결과를 새로운 변수에 할당합니다.
olderthan_19 = success_ch['char_age'] >= 19

# 두가지 조건를 동시에 충족하는 데이터를 필터링하여 새로운 변수에 저장합니다. (AND)
date_diff19 = success_ch[successwithin_19 & olderthan_19]

# 결과를 출력합니다.
date_diff19
```

```python
#21일 기준에서 성공프로그램 수, 배제되는 프로그램 수
# date-diff 컬럼을 선택
# 컬럼의 값과 조건을 비교
# 그 결과를 새로운 변수에 할당
successwithin_21 = success_ch['date_diff'] <= 21


# char_age 컬럼을 선택합니다.
# 컬럼의 값과 조건을 비교합니다.
# 그 결과를 새로운 변수에 할당합니다.
olderthan_21 = success_ch['char_age'] >= 21

# 두가지 조건를 동시에 충족하는 데이터를 필터링하여 새로운 변수에 저장합니다. (AND)
date_diff21 = success_ch[successwithin_21 & olderthan_21]

# 결과를 출력합니다.
date_diff21
```
